```python
def std_table(data, category, label):
    # Categorias ordenadas de mayor a menor por desvio standard en sus ventas
    category_group = data.groupby(category)[label].describe()
    print("ANALISIS ESTADISTICO")
    print("Categorias ordenadas de mayor a menor por desvio standard según label seleccionada")
    return category_group.sort_values(by="std", ascending=False)

std_table(ventas, 'Sede', 'Ventas')
```

<b>count:</b> cantidad de ventas en el período <br />
<b>mean:</b> promedio de ventas en el período <br />
<b>std:</b> desvío estandar de las ventas realizadas en el período (ordenada de mayor a menor) <br />
<b>min, 25%, 50%, 75%, max:</b> cuartiles <br />
